```python
import pandas as pd
import numpy as np
```

```python
file_path = '/content/TripB01.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1', delimiter=';')
selected_columns = [
    "Battery Voltage [V]",
    "Battery Current [A]",
    "Battery Temperature [°C]",
    "Ambient Temperature [°C]",
    "SoC [%]"
]
dataset = df[selected_columns]

print(dataset.head())
```

```python
dataset.describe()
null_values_column = df['Battery Voltage [V]'].isnull().sum()
print(null_values_column)
null_values_column = df['Battery Current [A]'].isnull().sum()
print(null_values_column)

null_values_column = df['Battery Temperature [°C]'].isnull().sum()
print(null_values_column)

null_values_column = df['Ambient Temperature [°C]'].isnull().sum()
print(null_values_column)
```

```python
from sklearn.model_selection import train_test_split
x = dataset.drop('SoC [%]', axis=1)
y = dataset['SoC [%]']
x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.2,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(
    n_estimators=75,
    random_state=42
)

rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
```
